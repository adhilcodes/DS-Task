```python
df1_copy = df1
```

```python
print(df1_copy.shape)
df1_copy.head()
```

**Adding new columns:**

 (1) ValidFileNames  	- List of all valid file names

 (2) InvalidFileNames	- List of all invalid file names

```python
%%time

df1_copy['ValidFileNames'] = ""
df1_copy['InvalidFileNames'] = ""
```

```python
print(df1_copy.shape)
df1_copy.head()
```

**Contracts are considered INVALID below conditions satisfied:**

(1) IsExecuted == False

(2) QualityScore < 0.81

```python
%%time


for ind in df1_copy.index:

  contract = df1_copy["FileName"][ind]

  if (df1_copy["IsExecuted"][ind]) == False or (df1_copy["QualityScore"][ind]) < 0.81:
    df1_copy["InvalidFileNames"][ind] = contract

  else:
    df1_copy["ValidFileNames"][ind] = contract
```

```python
print(df1_copy.shape)
df1_copy.head()
```

```python
%%time


# Deleting unwanted columns
df = df1.drop(['FileName', 'Language', 'StartDate', 'DocumentType', 'IsExecuted', 'QualityScore'], axis=1)
```

```python
print(df.shape)
df.head()
```

```python
%%time


# Assigning Labels to df

for i in df["CaseId"]:
  for j in df2["CaseId"]:

    if i == j:
      
      df["label_1"] = df2["label_1"]
      df["label_2"] = df2["label_2"]
```

```python
print(df.shape)
df.head()
```

```python
# Saving final prepared dataset : " df_final.gzip "


df.to_pickle('df_final.gzip')
```
